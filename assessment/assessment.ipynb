```python
import pandas as pd
import numpy as np
import datetime
pd.options.display.max_rows = 999
```

# Read data files

```python
case = pd.read_csv('data/case.csv')
demo = pd.read_csv('data/demo.csv')
hist = pd.read_csv('data/prior_arrests.csv') #hist reprensets arrest history, which is "prior_arrests" dataset
```

```python
case
```

# a. Create 're_arrest' variable 

```python
#sort the case dataframe by person_id
comb= case.sort(columns='person_id').head(n=100).reset_index(drop=True)
#Create 're_arrest' variable
re_arrest = []
for index, row in comb.iterrows():
    if index == 0:
        re_arrest.append(0)
    else:
        row1 = comb.irow(index-1)
        row2 = comb.irow(index)
        if row2.person_id == row1.person_id:
            if row2.arrest_date > row1.dispos_date:
                re_arrest.append(1)
            else:
                re_arrest.append(0)
        else:
            re_arrest.append(0)
            
comb['re_arrest'] = re_arrest
```

# b. Create 'prior_arrests' variable

```python
count_arrest = hist.groupby(['person_id'],as_index=False).count()
count_arrest = count_arrest.rename(columns={'arrest_date':'count_arrest'})
comb = case.merge(count_arrest, on='person_id',how='outer')
```

```python
prior_arrests = []
for index, row in comb.iterrows():
    
    if index == 0:
        count = comb.irow(index).count_arrest
        prior_arrests.append(count)
    else:
        row1 = comb.irow(index-1)
        row2 = comb.irow(index)
        if row2.person_id == row1.person_id:
            count = count + 1      
        else:
            count = comb.irow(index).count_arrest
        prior_arrests.append(count)
        
comb['prior_arrests'] = prior_arrests
comb['prior_arrests'].fillna(0, inplace=True)
```

# c.Create 'age' variable

```python
comb = comb.merge(demo, on='person_id',how='outer')
comb.head()
```

```python
comb['arrest_date'] = pd.to_datetime(comb['arrest_date'])
comb['dispos_date'] = pd.to_datetime(comb['dispos_date'])
comb['bdate'] = pd.to_datetime(comb['bdate'])
comb['age'] = comb['arrest_date'].dt.year - comb['bdate'].dt.year
```
